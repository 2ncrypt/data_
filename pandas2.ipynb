```python
robots = [[24, 23680],[35,NaN],[46, 47350],[27, NaN]]
data = pd.DataFrame(robots,columns=['max_speed','prince'])
display(data)
```

```python
data.dropna(subset=['prince'],axis=0, inplace=True)
display(data)
```

    Numpy에서 NaN을 가지고 와서 robots 리스트를 작성, 
    준비된 리스트를 pandas로 변환 후
    dropna() 메소드를 활용하여(NaN을 버려라)
    axis=0 행 삭제
    inplace=True : 해당 데이터 세트 자첼ㄹ 수정

```python
robots = [[24, 23680],[35,NaN],[46, 47350],[27, NaN]]
df = pd.DataFrame(robots,columns=['max_speed','price'])
mean = df.price.mean ##최솟값
df = df.replace(NaN, mean)
display(df)
```

```python
import pandas as pd
ary = [[2, 2], [4, 4], [6, 6], [8, 8], [10, 10]]
data = pd.DataFrame(ary, columns=['수온', '상온']) 
data['수온'] = data['수온']/data['수온'].max() 
data['상온'] = data['상온']/data['상온'].max()
```

```python
ary = [[1, 1.1 , '손'],[2, 2.2, '날개'],[3, 3.3, '손']]
data = pd.DataFrame(ary, columns=['수온', '상온', 'hand'])
data_0 = pd.get_dummies(data=['hand'])
display(data_0)
```

get_dummies() 문자형의 데이터를 숫자 형태로 변환하는 메소드

```python
display(data)
```

```python
data_1 = pd.get_dummies(data.hand)
display(data_1)
```
